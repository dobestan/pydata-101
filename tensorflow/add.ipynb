```python
import tensorflow as tf
```

```python
a = tf.constant(10)
b = tf.constant(20)
c = a + b
```

```python
c
```

```python
with tf.Session() as sess:
    result = sess.run(c)
    print(result)
```

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
c = a + b
```

```python
c
```

```python
with tf.Session() as sess:
    result = sess.run(c, feed_dict={a: 10, b: 20})
    print(result)
```
