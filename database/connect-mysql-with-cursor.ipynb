# MySQL 데이터베이스에 연결하기 - via cursor object

```python
import os
```

```python
import pandas as pd

import MySQLdb
```

```python
DATABASE_HOST = "localhost"  # "localhost" == "127.0.0.1"

DATABASE_USERNAME = os.environ.get("DATABASE_USERNAME", "YOUR_USERNAME")
DATABASE_PASSWORD = os.environ.get("DATABASE_PASSWORD", "YOUR_PASSWORD")

DATABASE_NAME = "sakila"
```

```python
connection = MySQLdb.connect(
    DATABASE_HOST,
    DATABASE_USERNAME,
    DATABASE_PASSWORD,
    DATABASE_NAME,
    charset='utf8',
)
```

```python
# 데이터베이스에 연결하고 난 이후 ( "connection" 객체를 생성하고 난 이후, )
# cursor 객체를 생성해주셔야 합니다.


cursor = connection.cursor()
```

```python
SQL_QUERY = """
    SELECT *
    FROM customer
    ;
"""

cursor.execute(SQL_QUERY)
cursor.fetchall()
```

```python
# 다만, 이 경우에는 column에 대한 정보를 가져오지 못한다.
# ( 이럴 경우는 잘 없겠지만, ) 데이터베이스의 구조에 대해서 잘 모른다면,
# 다음의 SQL 쿼리를 통해서 Column Name 에 대한 정보를 얻을 수 있다.


SQL_QUERY = """
    SELECT column_name
    FROM information_schema.columns
    WHERE table_name='customer'
"""

cursor.execute(SQL_QUERY)
cursor.fetchall()
```
